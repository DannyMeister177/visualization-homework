### Homework 6: Exploring project dataset

Start with setup...

```python
import pandas as pd
import matplotlib.pyplot as plt
import os
%matplotlib inline

# Load the data
df = pd.read_csv('data/Project_data/WA_Fn-UseC_-HR-Employee-Attrition.csv')
df.drop('EmployeeCount', axis=1, inplace=True)        # Column has no relevant info

os.makedirs('plots/reach_jupyter-notebook', exist_ok=True)
```

Line Plots

```python
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
```

```python
# Age
age_sorted = df['Age'].sort_values().reset_index(drop=True)

axes[0][0].plot(age_sorted, label='Age', color='g')
axes[0][0].set_title('Age Line Plot')
axes[0][0].set_xlabel('Index')
axes[0][0].set_ylabel('Age')

# add mean Age
age_mean = age_sorted.mean()
axes[0][0].axhline(age_mean, linestyle='--', color='#147FB0')
axes[0][0].text(10, age_mean+1.1, f'Mean = {age_mean:.0f}', color='#147FB0')

# Annual Income
annual_income = (df['MonthlyIncome']*12).sort_values().reset_index(drop=True)

axes[0][1].plot(annual_income, label='Annual Income', color='g')
axes[0][1].set_title('Annual Income Line Plot')
axes[0][1].set_xlabel('Index')
axes[0][1].set_ylabel('Annual Income')

# add median Annual Income
ai_median = annual_income.median()
axes[0][1].axhline(ai_median, linestyle='--', color='#147FB0')
axes[0][1].text(10, ai_median+10000, f'Median = $ {ai_median:.2f}', color='#147FB0')

# YearsAtCompany
yac_sorted = df['YearsAtCompany'].sort_values().reset_index(drop=True)

axes[1][0].plot(yac_sorted, label='Years At Company', color='g')
axes[1][0].set_title('Years At Company Line Plot')
axes[1][0].set_xlabel('Index')
axes[1][0].set_ylabel('Years At Company')

# add median YearsAtCompany
yac_median = yac_sorted.median()
axes[1][0].axhline(yac_median, linestyle='--', color='#147FB0')
axes[1][0].text(10, yac_median+1.1, f'Median = {yac_median:.0f}', color='#147FB0')

# YearsWithCurrManager
ywcm_sorted = df['YearsWithCurrManager'].sort_values().reset_index(drop=True)

axes[1][1].plot(ywcm_sorted, label='Years with Current Manager', color='g')
axes[1][1].set_title('Years with Current Manager Line Plot')
axes[1][1].set_xlabel('Index')
axes[1][1].set_ylabel('Years with Current Manager')

# add median YearsWithCurrManager
ywcm_median = ywcm_sorted.median()
axes[1][1].axhline(ywcm_median, linestyle='--', color='#147FB0')
axes[1][1].text(10, ywcm_median+0.5, f'Median = {ywcm_median:.0f}', color='#147FB0')

fig.savefig('plots/reach_jupyter-notebook/attrition_line_plots.png', dpi=300)

fig
```

Histograms

```python
# Age
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes[0][0].hist(df['Age'], color='g')
axes[0][0].set_title('Age Histogram')
axes[0][0].set_xlabel('Buckets')

# Annual income
axes[0][1].hist(annual_income, color='g')
axes[0][1].set_title('Annual Income Histogram')
axes[0][1].set_xlabel('Buckets')

# YearsAtCompany
axes[1][0].hist(df['YearsAtCompany'], color='g')
axes[1][0].set_title('Years at Company Histogram')
axes[1][0].set_xlabel('Buckets')

# YearsWithCurrManager
axes[1][1].hist(df['YearsWithCurrManager'], color='g')
axes[1][1].set_title('Years with Current Manager Histogram')
axes[1][1].set_xlabel('Buckets')

fig.savefig('plots/reach_jupyter-notebook/attrition_histogram_plots.png', dpi=300)
```

Scatter plots

```python
fig, axes = plt.subplots(2, 2, figsize=(13, 10))

# Age to Annual Income
axes[0][0].scatter(df['Age'], df['MonthlyIncome']*12, s=5, label='Age vs Annual Income', 
                   color='purple', marker='^')
axes[0][0].set_title('Age vs Income')
axes[0][0].set_xlabel('Age')
axes[0][0].set_ylabel('Annual Income')
axes[0][0].legend()

# Gender to Annual Income
axes[0][1].scatter(df['Gender'], df['MonthlyIncome']*12, s=5, label='Gender vs Annual Income', 
                   color='purple', marker='^')
axes[0][1].set_title('Gender vs Income')
axes[0][1].set_xlabel('Gender')
axes[0][1].set_ylabel('Annual Income')
axes[0][1].legend()


# MonthlyIncome to MonthlyRate
axes[1][0].scatter(df['MonthlyIncome'], df['MonthlyRate'], s=5, label='MonthlyIncome vs MonthlyRate', 
                   color='purple', marker='^')
axes[1][0].set_title('MonthlyIncome vs MonthlyRate')
axes[1][0].set_xlabel('MonthlyIncome')
axes[1][0].set_ylabel('MonthlyRate')
axes[1][0].legend()


# Age to YearsAtCompany to YearsSinceLastPromotion
axes[1][1].scatter(df['YearsAtCompany'], df['YearsSinceLastPromotion'], s=5, color='purple', marker='^')
axes[1][1].set_title('Years at Comp vs Years since Promoted')
axes[1][1].set_xlabel('YearsAtCompany')
axes[1][1].set_ylabel('YearsSinceLastPromotion')
axes[1][1].legend()

fig.savefig('plots/reach_jupyter-notebook/attrition_scatter_plots.png', dpi=300)
```
